```python
import folium
import math

# --- Helper: haversine distance in kilometers ---
def haversine_km(lat1, lon1, lat2, lon2):
    R = 6371.0
    p1, p2 = math.radians(lat1), math.radians(lat2)
    dphi = math.radians(lat2 - lat1)
    dlmb = math.radians(lon2 - lon1)
    a = (math.sin(dphi/2)**2
         + math.cos(p1)*math.cos(p2)*math.sin(dlmb/2)**2)
    return 2 * R * math.asin(math.sqrt(a))

# --- Coordinates (example) ---
site_name = "CCAFS LC-40"
site = (28.5618571, -80.577366)

# Replace these with actual points you identify on the map
railway = (28.5629, -80.8074)   # EXAMPLE railway coord
highway = (28.5692, -80.6558)   # EXAMPLE highway ramp/exit
coast   = (28.6000, -80.5400)   # EXAMPLE coastline point

features = {
    "Nearby Railway": railway,
    "Highway Access": highway,
    "Coastline": coast
}

# --- Build map ---
m = folium.Map(location=site, zoom_start=10)

# Launch site marker
folium.Marker(
    site, popup=f"{site_name}",
    icon=folium.Icon(color="blue", icon="rocket", prefix="fa")
).add_to(m)

# Feature markers + lines + distance labels
for label, (lat, lon) in features.items():
    d_km = haversine_km(site[0], site[1], lat, lon)
    folium.Marker(
        (lat, lon),
        popup=f"{label}<br>Distance: {d_km:.2f} km",
        icon=folium.Icon(color="green" if label!="Coastline" else "cadetblue")
    ).add_to(m)

    # Polyline from site to feature with tooltip distance
    folium.PolyLine(
        [site, (lat, lon)],
        tooltip=f"{label}: {d_km:.2f} km",
        weight=3
    ).add_to(m)

# Optional: circle around site (e.g., 5km radius)
folium.Circle(
    location=site, radius=5000, fill=True, fill_opacity=0.05,
    popup="5 km radius"
).add_to(m)

m.save("launch_site_proximity.html")
print("Saved: launch_site_proximity.html  → open in your browser and take a screenshot.")
```
